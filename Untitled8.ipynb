```python
from sklearn import linear_model
```

```python
house=pd.read_csv('housing.csv')
```

```python
house.head()
```

```python
x=house.drop('MEDV',axis=1).values
y=house['MEDV'].values
```

```python
x_rooms=x[:,1]
```

```python
x_rooms=x_rooms.reshape(-1,1)
y=y.reshape(-1,1)
```

```python
x_rooms.shape
```

```python
plt.scatter(x_rooms,y)
```

```python
reg=linear_model.LinearRegression()
```

```python
reg.fit(x_rooms,y)
```

```python
from sklearn.metrics import mean_absolute_error,mean_squared_error,r2_score
y_predict=reg.predict(x_rooms)
```

```python
print(mean_absolute_error(y,y_predict))
```

```python
print(r2_score(y,y_predict))
```

```python
reg.fit(x,y)
```

```python
from sklearn.metrics import mean_absolute_error,mean_squared_error,r2_score
y_predict=reg.predict(x)
```

```python
print(mean_absolute_error(y,y_predict))
```

```python
print(r2_score(y,y_predict))
```
